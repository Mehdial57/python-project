```python
str1 = 'ایران‌ خودرو'
str2 = 'ایران‌ خودرو'
def compare_strings(str1, str2):
    # Find the minimum length between the two strings
    min_length = min(len(str1), len(str2))

    # Initialize an empty list to store the differing characters
    differing_chars = []

    # Compare each character in the strings
    for i in range(min_length):
        if str1[i] != str2[i]:
            differing_chars.append((str1[i], str2[i]))
#If one string is longer than the other, add the extra characters
    if len(str1) > len(str2):
        differing_chars.extend([(char, None) for char in str1[min_length:]])
    elif len(str1) < len(str2):
        differing_chars.extend([(None, char) for char in str2[min_length:]])

    return differing_chars
differences = compare_strings(str1, str2)
print(differences)
```

```python
from fuzzywuzzy import fuzz
```

```python
def merge_dataframes(df1, df2, threshold):
    merged_df = pd.DataFrame()
    for i, row in df1.iterrows():
        best_match = max(df2['نام شرکت'].apply(lambda x: fuzz.ratio(row['نام شرکت'], x)), default=0)
        if best_match >= threshold:
            merged_df = pd.concat([merged_df, pd.concat([row.to_frame().T, df2[df2['نام شرکت'] == df2['نام شرکت'][best_match]].iloc[0].to_frame().T], axis=1)], ignore_index=True)
    return merged_df
```

```python
merged_df =merge_dataframes(company_list, df, 80)
```
