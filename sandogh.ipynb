```python
import pandas as pd
```

```python
df = pd.read_excel('ثهام-اسفند1402.xlsx', sheet_name='سهام')
```

```python
df = df.iloc[6:].reset_index(drop=True)
```

```python
df.columns = df.iloc[0]
```

```python
df.dropna(axis='columns', how='all', inplace=True)
```

```python
df = df.iloc[1:]
```

```python
df
```

```python
df = df.loc[:, ~df.columns.duplicated(keep='last')]
```

```python
df = df.drop(columns={'مبلغ فروش', 'خالص ارزش فروش'}, axis=1)
```

```python
df['تعداد'] = pd.to_numeric(df['تعداد'])
df['بهای تمام شده'] = pd.to_numeric(df['بهای تمام شده'])
```

```python
df['میانگین قیمت خرید'] = (df['بهای تمام شده'] / df['تعداد']).round(2)
```

```python
df['درصد به کل دارایی‌های صندوق'] = (df['بهای تمام شده'] / df['بهای تمام شده'].sum() * 100).round(2)
```

```python
df.to_excel('ثهام.xlsx', index=False)
```
